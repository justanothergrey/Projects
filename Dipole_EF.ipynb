### Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define grid
Aquí definimos cómo será el espacio de nuestro dipolo

```python
#Generamos un espacio de -11 a 11 con 50 ítems
x = np.linspace(-11,11,50) 
y = np.linspace(-11,11,50)
X,Y = np.meshgrid(x,y) 
```

### Define dipole parameters
Aquí definimos las cargas y sus distancias

```python
q1 = 1
q2 = -q1
r1x = -5
r1y = 0
r2x = 5
r2y = 0

r13 = np.sqrt((X-r1x/2)**2 + (Y-r1y)**2)
r23 = np.sqrt((X-r2x/2)**2 + (Y-r2y)**2)

#r1 = np.sqrt((X - a / 2)**2 + Y**2)
#r2 = np.sqrt((X + a / 2)**2 + Y**2)
```

### Define Electric field conditions
Aquí definimos las condiciones del campo eléctrico

```python
#Calculamos las atribuciones al campo
E1x = q1*(X-r1x)/r13**3
E2x = q2*(X-r2x)/r23**3
E1y = q1*(Y-r1y)/r13**3
E2y = q2*(Y-r2y)/r23**3

#Finalmente, calculamos con principio de superposición
Ex = E1x + E2x
Ey = E1y + E2y
```

### Plot Electric field of a dipole
Ahora graficamos el campo eléctrico del dipolo.

```python
plt.figure()
plt.streamplot(X,Y,Ex,Ey)
plt.title('Campo Eléctrico de un Dipolo (streamplot)')
plt.show()

plt.figure()
plt.quiver(X,Y,Ex,Ey,scale=60)
plt.title('Campo Eléctrico de un Dipolo (Quiver)')
plt.show()
```
