```python
#Imports y parámetros para graficar
import numpy as np 
import matplotlib.pyplot as plt
t = np.linspace (0, 2, 1000)
l1 = 10
l2 = 50
N1i = 1
N2i = 0
N3i = 0
```

```python
#Expresion para cada tipo de nucleo en un tiempo t
N1 = N1i * np. exp (-l1 * t)
N2 = N2i + N1i/l1 * (np.exp(-l1 * t) - np.exp (-l2 * t))
N3 = N3i + N1i/l1 * (1 - np.exp (-l1 * t)) + N2i/l2 * (1 - np.exp (-l2 * t))
```

```python
#Ahora calculamos las abundancias relativas en funcion de tiempo
Nt= N1 + N2 + N3
x1 = N1 / Nt
x2 = N2 / Nt
x3 = N3 / Nt
```

```python
#Graficamos
plt.plot (t, x1, label= 'Type 1')
plt.plot (t, x2, label= 'Type 2')
plt.plot (t, x3, label= 'Type 3')
plt.xlabel( 'Tiempo (s)')
plt.ylabel ( 'Abundancia Relativa')
plt.show ( )
```
